```python
freq_index_1 = (np.abs(output_data['calc_data']['results']['f_sample_values']- 10.0)).argmin()
freq_index_2 = (np.abs(output_data['calc_data']['results']['f_sample_values']- 100.0)).argmin()
freq_index_3 = (np.abs(output_data['calc_data']['results']['f_sample_values']- 1000.0)).argmin()
freq_index_4 = (np.abs(output_data['calc_data']['results']['f_sample_values']- 10000.0)).argmin()



print(output_data['calc_data']['results']['units'])
for i in range(len(output_data['calc_data']['m_wimp'])):
    plt.plot(output_data['calc_data']['results']['ang_sample_values'],np.array(output_data['calc_data']['results']['final_data'][i][freq_index_1])/11818102.86,label=r"$M_\chi$ = "+str(output_data['calc_data']['m_wimp'])+" GeV")
    plt.plot(output_data['calc_data']['results']['ang_sample_values'],np.array(output_data['calc_data']['results']['final_data'][i][freq_index_2])/11818102.86,label=r"$M_\chi$ = "+str(output_data['calc_data']['m_wimp'])+" GeV")
    plt.plot(output_data['calc_data']['results']['ang_sample_values'],np.array(output_data['calc_data']['results']['final_data'][i][freq_index_3])/11818102.86,label=r"$M_\chi$ = "+str(output_data['calc_data']['m_wimp'])+" GeV")
    plt.plot(output_data['calc_data']['results']['ang_sample_values'],np.array(output_data['calc_data']['results']['final_data'][i][freq_index_4])/11818102.86,label=r"$M_\chi$ = "+str(output_data['calc_data']['m_wimp'])+" GeV")
plt.xscale("log")
plt.yscale("log")
plt.ylabel(r"I$(\nu)$ (Jy/arcmin$^2$)",fontsize=16)
plt.xlabel(r"$\Theta$ (arcmins)",fontsize=16)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.title(f"Frequency = {output_data['calc_data']['results']['f_sample_values'][freq_index_1]:.2f} MHz")
plt.title(f"Frequency = {output_data['calc_data']['results']['f_sample_values'][freq_index_2]:.2f} MHz")
plt.title(f"Frequency = {output_data['calc_data']['results']['f_sample_values'][freq_index_3]:.2f} MHz")
plt.title(f"Frequency = {output_data['calc_data']['results']['f_sample_values'][freq_index_4]:.2f} MHz")
plt.legend()
plt.show()
```
